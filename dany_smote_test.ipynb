```python
X = data.drop(columns='Target')
y = data['Target']
```

```python
from imblearn.over_sampling import SMOTE
```

```python
over = SMOTE(sampling_strategy=0.3, random_state=1)
```

```python
X, y = over.fit_resample(X, y)
```

```python
len(X)
```

```python
y.value_counts()
```
