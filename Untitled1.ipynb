```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/data_startup_final.csv')
df
```

```python
df.count()
```

```python
df['last_funding_at'] = pd.to_datetime(df['last_funding_at'])
df['exited_on'] = pd.to_datetime(df['exited_on'])
```

```python
last_funding_count = df.groupby(df['last_funding_at'].dt.year)['last_funding_at'].count()
exited_on_count = df.groupby(df['exited_on'].dt.year)['exited_on'].count()
```

```python
plt.figure(figsize=(8, 8))
plt.bar(last_funding_count.index, last_funding_count.values, width=0.4, label='Last Funding At')
plt.bar(exited_on_count.index + 0.4, exited_on_count.values, width=0.4, label='Exited On')
plt.xticks(last_funding_count.index)
plt.xlim([2015, 2024])
plt.xlabel('Year')
plt.ylabel('Company Count')
plt.title ('Last Funding Round vs Exit Year')
plt.legend()
plt.show()
```

```python
df = pd.read_csv('data/data_startup_final.csv')
```

```python
df.count()
```

```python
grouped = df.groupby('name').agg({'last_funding_at': lambda x: pd.to_datetime(x).mean(),
                                  'exited_on': lambda x: pd.to_datetime(x).mean()})
mean_last_funding_at = grouped['last_funding_at']
mean_exited_on = grouped['exited_on']
```

```python
plt.plot(mean_last_funding_at, color='blue', label='Mean last_funding_at')
plt.plot(mean_exited_on, color='red', label='Mean exited_on')
plt.xlabel('Company')
plt.ylabel('Date')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```
